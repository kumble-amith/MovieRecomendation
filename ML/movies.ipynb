```python
import pandas as pd
import numpy as np 
import ast
```

```python
def convert(value):
    X = []
    for i in ast.literal_eval(value):
        X.append(i['name'].replace(" ",''))
    return X

def convertOverview(value):
    return value.split(' ')
```

```python
def convert3(value):
    X = []
    count = 0
    for i in ast.literal_eval(value):
        X.append(i['name'].replace(" ",''))
        count += 1
        if count == 3:
            break
    return X

def getDirector(value):
    director = ast.literal_eval(value)
    X = []
    for i in director:
        if i['job'].lower() == 'director':
            X.append(i['name'].replace(" ",''))
            break
    return X
```

```python
movies = pd.read_csv("Movies.csv",converters={
    'genres':convert,
    'keywords':convert,
    'overview':convertOverview
})
credit = pd.read_csv("Credits.csv",converters={
    'cast':convert3,
    'crew':getDirector
})
```

```python
movies = movies.merge(credit,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x).lower())
```

```python
data = movies[['movie_id','title','tags']]
```

```python
data
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vectors)
```

```python
def recomend(movie):
    index = data[data['title'] == movie].index[0]
    dist = similar[index]
    new_list = sorted(list(enumerate(dist)),reverse =True ,key = lambda x :x[1])[1:6]
    for i in new_list:
        print(data.iloc[i[0]].title)
```

```python
recomend('Batman')
```
